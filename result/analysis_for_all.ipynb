```python
import os, json

def read_json_file(filePath):
    with open(filePath) as f:
        data = json.load(f)
    return data
```

```python
filepath = 'jsondata/'
raw_data = {}
for i in sorted(os.listdir(filepath)):
    if '.ipy' in i or '.DS' in i:
        continue
    temp_data = read_json_file(os.path.join(filepath, i))
    temp_exp = i[:-10]
    # print(temp_data)
    # reconstruct data
    temp_reconstruct_data = {}
    for j in temp_data:
        # print(j)
        temp_reconstruct_data[j[0]] = j[1]
        # break
    # print(temp_reconstruct_data[3])
    # break
    raw_data[temp_exp] = temp_reconstruct_data
    
```

```python
# Compute Rank error
def get_average_rank(result ,exp_name_rank_order_pyfeat):
    aver_ranking = list(exp_name_rank_order_pyfeat.items())
    for i, c in enumerate(aver_ranking):
        aver_ranking[i] = [c[0], 0]

    for i in range(len(result)):
        # print(result)

        temp_data = list(enumerate(list(result.items())[i][1]))
        
        temp_list = []
        for i, c in enumerate(temp_data):
            temp_list.append((c, exp_name_rank_order_pyfeat[c[1]]))
        temp_list.sort(key=lambda x:x[1])
        # print(temp_list)
        for i, c in enumerate(aver_ranking):
            if c[0] == temp_list[i][0][1]:
                aver_ranking[i][1] += temp_list[i][0][0]
    aver_ranking_new = list(enumerate(aver_ranking))
    aver_ranking_new.sort(key=lambda x:x[1][1], reverse=False)
    output_list = [kk[1][0] for kk in aver_ranking_new]
    return output_list, aver_ranking_new

# exp angry
pier_dic = {}

for exp,v in raw_data.items():
    print(exp)
    subject_data = raw_data[exp]
    feat_data = feat_order[exp]
    feat_data = {v:idx for idx,v in enumerate(feat_data)}

    average_rank, _ = get_average_rank(subject_data, feat_data)
    print(average_rank)
    # print(list(subject_data.items()))

    pier_score = []
    # print(subject_data)
    for idx, v in subject_data.items():
        temp_list = []
        tmp_ll = list(enumerate(v))
        # print(tmp_ll)
        for j, c in enumerate(tmp_ll):
            # print(c)
            temp_list.append((c, feat_data[c[1]]))
        temp_list.sort(key=lambda x:x[1])

        x = [k[1] for k in temp_list]
        y = []
        temp_score = 0
        for k in temp_list:
            tmp = abs(k[1] - k[0][0]) ** 2
            temp_score += tmp
            y.append(temp_score)
        pier = y[-1]
        if idx != 6:
            # pier_score.append([idx, pier])
            pier_score.append(pier)
        # print(pier)
        
    # average rank
    temp_list = []
    tmp_ll = list(enumerate(average_rank))
    for j, c in enumerate(tmp_ll):
        # print(j, c)

        temp_list.append((c, feat_data[c[1]]))
    temp_list.sort(key=lambda x:x[1])
    y = []
    temp_score = 0
    # print('temp_list', temp_list)
    for k in temp_list:
        tmp = abs(k[1] - k[0][0]) ** 2
        temp_score += tmp
        y.append(temp_score)
    pier = y[-1]
    # pier_score.append([len(subject_data.items()) + 1, pier])
    # pier_score.append([6, pier])
    pier_score.append(pier)
    
    pier_dic[exp] = pier_score
    # break
# temp_list = [x[0] for x in temp_list]
```
